```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('Go_Auto_Dataset.csv')
```

```python
df_active = df[df['listing_type'] == 'Active']
df_sold = df[df['listing_type'] == 'Sold']
```

```python
df_active_new = df_active[df_active['stock_type'] == 'NEW']
df_active_used = df_active[df_active['stock_type'] == 'USED']
```

```python
df_active_new['series'] = df_active_new['series'].fillna(df_active_new['series'].mode()[0])

df_active_new['exterior_color_category'] = df_active_new['exterior_color_category'].fillna(method = 'ffill')
df_active_new['exterior_color_category'] = df_active_new['exterior_color_category'].fillna(method = 'bfill')

df_active_new['interior_color_category'] = df_active_new['interior_color_category'].fillna(method = 'ffill')
df_active_new['interior_color_category'] = df_active_new['interior_color_category'].fillna(method = 'bfill')
```

```python
Q1 = df_active_new['days_on_market'].quantile(0.25)
Q3 = df_active_new['days_on_market'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


df_active_new['days_on_market'] = np.where(df_active_new['days_on_market'] > upper_bound, upper_bound, np.where(df_active_new['days_on_market'] < lower_bound, lower_bound, df_active_new['days_on_market']))
```

```python
Q1 = df_active_new['mileage'].quantile(0.25)
Q3 = df_active_new['mileage'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


df_active_new['mileage'] = np.where(df_active_new['mileage'] > upper_bound, upper_bound, np.where(df_active_new['mileage'] < lower_bound, lower_bound, df_active_new['mileage']))
```

```python
Q1 = df_active_new['price'].quantile(0.25)
Q3 = df_active_new['price'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1
upper_bound = Q3 + 1.5 * IQR


df_active_new['price'] = np.where(df_active_new['price'] > upper_bound, upper_bound, np.where(df_active_new['price'] < lower_bound, lower_bound, df_active_new['price']))
```

```python
Q1 = df_active_new['msrp'].quantile(0.25)
Q3 = df_active_new['msrp'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1
upper_bound = Q3 + 1.5 * IQR


df_active_new['msrp'] = np.where(df_active_new['msrp'] > upper_bound, upper_bound, np.where(df_active_new['msrp'] < lower_bound, lower_bound, df_active_new['msrp']))
```
